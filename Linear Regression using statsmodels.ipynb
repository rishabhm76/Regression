Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Open CSV File

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
```

```python
data
```

```python
data.describe()
```

Define variables

```python
y = data['GPA']
x1 = data['SAT']
```

Plot Data

```python
plt.scatter(x1,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

Analyse Data

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

Plot Least Square Regression

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat, lw=3, c='orange', label='regression line')
plt.xlabel('SAT', fontsize = 18)
plt.ylabel('GPA', fontsize = 18)
plt.show()
```
